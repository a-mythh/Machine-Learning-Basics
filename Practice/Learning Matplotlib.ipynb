#### Load necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Line Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Set figure size (dpi sets pixels per inch. Keep it at 300 for clearer images)
plt.figure(figsize = (5, 3), dpi = 120)

# Plot a line graph - Line 1
plt.plot(x, y, label = '2x', color = '#f220b3', linewidth = 1.5, linestyle = '--', marker = '.', markersize = 10, markeredgecolor = 'black')

# shorthand notation
# fmt = '[color][marker][linestyle]'
# plt.plot(x, y, 'yx--', label = '2x')

# Line 2
x2 = np.arange(0, 4.5, 0.5)

# plot line 2 - make it like a projection
plt.plot(x2[:6], x2[:6] ** 2, 'y-', label = 'x^2')
plt.plot(x2[5:], x2[5:] ** 2, 'y--')

# renaming the label of the axes
plt.xlabel("x-axis")
plt.ylabel('y-axis')

# set title for the graph
plt.title('My First Graph!', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 16})

# mark points on x and y axes
plt.xticks([1, 2, 3, 4])
# plt.yticks([2, 4, 6, 8, 10])

# show a legend
plt.legend()

# save the graph
plt.savefig('mygraph.png', dpi = 300)

plt.show()
```

### Bar Chart

```python
values = [1, 8, 3, 5, 3]
labels = ['A', 'B', 'C', 'D', 'E']

# plot the bar chart
bars = plt.bar(labels, values, color = 'aqua')

# set different hatch values
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
bars[3].set_hatch('-')
bars[4].set_hatch('.')

plt.show()
```
