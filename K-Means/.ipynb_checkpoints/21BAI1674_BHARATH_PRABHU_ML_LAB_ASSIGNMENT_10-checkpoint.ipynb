**USING CUSTIOM DEFINED FUNCTION:**

```python
# Extract the features (first two columns) for flame
X = flame_df.iloc[:, :2].values
```

```python
# Extract the features (first two columns) for toy
X1 = toy_df.iloc[:, :2].values
```

```python
# Custom K-means clustering implementation
def custom_kmeans(X, n_clusters, max_iter=100, seed=None):
    np.random.seed(seed)
    # Randomly initialize centroids
    centroids = X[np.random.choice(range(len(X)), size=n_clusters, replace=False)]

    # Iteratively update centroids and assign labels
    for _ in range(max_iter):
        # Assign labels to each point
        labels = np.argmin(np.linalg.norm(X[:, np.newaxis] - centroids, axis=-1), axis=-1)

        # Update centroids
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(n_clusters)])

        # Check convergence
        if np.allclose(centroids, new_centroids):
            break

        centroids = new_centroids

    return labels, centroids
```

```python
# Custom K-means clustering for flame
custom_labels, custom_centroids = custom_kmeans(X, n_clusters=2, max_iter=100, seed=42)
```

```python
# Plotting the clusters obtained (custom implementation) for flames
plt.scatter(X[custom_labels == 0, 0], X[custom_labels == 0, 1], color='red', label='Cluster 1')
plt.scatter(X[custom_labels == 1, 0], X[custom_labels == 1, 1], color='blue', label='Cluster 2')
plt.scatter(custom_centroids[:, 0], custom_centroids[:, 1], color='black', marker='X', s=100, label='Centroids')
plt.title('Clusters Obtained (Custom)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
# Custom K-means clustering for toy
custom_labels1, custom_centroids1 = custom_kmeans(X1, n_clusters=2, max_iter=100, seed=42)
```

```python
# Plotting the clusters obtained (custom implementation) for toy
plt.scatter(X1[custom_labels1 == 0, 0], X1[custom_labels1 == 0, 1], color='red', label='Cluster 1')
plt.scatter(X1[custom_labels1 == 1, 0], X1[custom_labels1 == 1, 1], color='blue', label='Cluster 2')
plt.scatter(custom_centroids1[:, 0], custom_centroids1[:, 1], color='black', marker='X', s=100, label='Centroids')
plt.title('Clusters Obtained (Custom)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['shop (1).csv']))
df
```

```python
# Extract the features (Age, Annual Income, Spending Score)
Y = df.iloc[:, 2:5]
```

```python
# Determine the optimal number of clusters using the elbow method
inertia = []
silhouette_scores = []
rand_scores = []
max_clusters = 10

for n_clusters in range(1, max_clusters+1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(Y)
    inertia.append(kmeans.inertia_)

    if n_clusters > 1:
        silhouette_scores.append(silhouette_score(Y, kmeans.labels_))
        rand_scores.append(adjusted_rand_score(df['cluster'], kmeans.labels_))
```

```python
# Plotting the elbow curve
plt.plot(range(1, max_clusters+1), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()
```

```python
# Determine the optimal number of clusters from the elbow curve
optimal_clusters = 5  # Update with the elbow curve analysis
```

```python
# Perform K-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(Y)
```

```python
# Get the cluster labels
labels = kmeans.labels_

# Add the cluster labels to the dataset
df['cluster'] = labels
```

```python
# Print the cluster centroids
print('Cluster Centroids:')
print(kmeans.cluster_centers_)
```

```python
# Plotting the clusters
plt.scatter(Y.iloc[:, 0], Y.iloc[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='X', label='Centroids')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.title('Customer Segmentation')
plt.legend()
plt.show()
```

```python
# Plotting the Silhouette index scores
plt.plot(range(2, max_clusters+1), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Index')
plt.title('Silhouette Index')
plt.show()
```

```python
# Plotting the Rand index scores
plt.plot(range(2, max_clusters+1), rand_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Rand Index')
plt.title('Rand Index')
plt.show()
```
