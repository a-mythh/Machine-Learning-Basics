## K-Means Algorithm Implementation

### Flame dataset

#### Load necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

#### Load the dataset

```python
data = pd.read_csv('flame.csv')

data.head()
```

#### Split data into features and target

```python
x = data.drop('y', axis = 1)
y = data.y

x[:5]
```

### Plot data on 2D graph

```python
plt.xlabel('x1')
plt.ylabel('x2')

plt.scatter(x.x1, x.x2)

plt.show()
```

```python
x_array = np.array(x)
```

### K-Means Clustering using Sklearn

#### Load the libraries

```python
from sklearn.cluster import KMeans
```

#### Create instance of the model

```python
kmeans = KMeans(n_clusters = 2, random_state = 0, n_init = 'auto')
```

#### Fit the data into the model

```python
kmeans.fit(x_array)
```

#### Find the centres

```python
centers = kmeans.cluster_centers_

print(centers)
```

### Plot the clusters

```python
plt.scatter(x.x1, x.x2, c = kmeans.labels_)
plt.plot(centers[0][0], centers[0][1], 'co', 
         centers[1][0], centers[1][1], 'mo')

plt.show()
```

### Create a different instance of the model

```python
kmeans2 = KMeans(n_clusters = 2,  init='k-means++', random_state = 0, n_init = 20, max_iter = 700)
```

### Fit the data into the model

```python
kmeans2.fit(x_array)
```

```python
centers = kmeans2.cluster_centers_
```

### Plot the clusters

```python
plt.scatter(x.x1, x.x2, c = kmeans2.labels_)
plt.plot(centers[0][0], centers[0][1], 'ro', 
         centers[1][0], centers[1][1], 'mo')

plt.show()
```

***We can see that there are some changes in the clusters once we change the parameters. These can be tested and played with.***
